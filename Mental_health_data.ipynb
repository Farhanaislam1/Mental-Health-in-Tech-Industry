### Here, we can see that the gender column is made up of multiple different answers which will make it difficult for us to group them the answers together. Therefore, we need to categorize them manually in three categories (male, female and other).

```python
#Now we can Define a function to categorize gender
def categorize_gender(gender):
    if pd.isnull(gender):
        return "Other"
    gender = gender.lower()
    if gender in ['M', 'Male', 'male', 'm', 'Male-ish', 'maile', 'Cis Male', 'Mal'
 'Male (CIS)', 'Make', 'Male','Man', 'Mail', 'cis male', 'Malr','Cis Man']:
        return "Male"
    elif gender in ['Female', 'female', 'Trans-female', 'Cis Female', 'F', 'Woman', 'f', 'Femake', 'woman', 'Female ',
 'cis-female/femme', 'Trans woman','Female (trans)','Female (cis)', 'femail']:
        return "Female"
    else:
        return "Other"

```

```python
df['Gender_Categorized'] = df['Gender'].apply(categorize_gender) # # Now Apply the categorization function to the Gender column
df['Gender_Categorized']
```

### Clean age column


```python
import pandas as pd
import numpy as np
df = pd.read_csv("survey.csv")
Age = df['Age']
np.mean(Age)
```

## The mean value of age is unrealistic. so we have to filter age column from 19 to 80 to reaplace a realistic age value for the rest of the age value

```python
Age = df['Age']

upper_limit = 80
lower_limit = 19

mean = np.mean(Age[(Age > lower_limit) & (Age < upper_limit)])
mean = round(mean, 0) #rounding our decimals
mean
```

### Let use age 32 as our median.

### Now we can replace those ages that are lower than 19 and higher than 80 to our mean value of age 32.

```python
# Replace elements below the lower_limit with the calculated mean
df.loc[df['Age'] < lower_limit, 'Age'] = mean
# Replace elements above the upper_limit with the calculated mean
df.loc[df['Age'] > upper_limit, 'Age'] = mean

unique_age_values = df['Age'].unique()
# Print the unique age values
print(unique_age_values)
```

## we can see all the ages are fall between 19 to 80

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a boxplot for Age grouped by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Gender_Categorized", y="Age", palette="Set2")
plt.title("Age Distribution by Gender (Boxplot)", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

here we can see the distribution of genders and the groups of ages in our dataset. We can conclude that our dataset consist mostly of male in their late 20s to their early 30s.

### To analyze the relationship between age and treatment, we can use visualizations like boxplots, histograms, and scatter plots, grouped or color-coded by the treatment status. Here's how to do this in Python:

```python
# Boxplot: Age distribution by Treatment
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="treatment", y="Age", palette="Set2")
plt.title("Age Distribution by Treatment", fontsize=16)
plt.xlabel("Treatment", fontsize=14)
plt.ylabel("Age", fontsize=14)
plt.show()
```

 the relationship between the subjects that got treatment and their ages.From this visualization, we can conclude that people who are more likely to not get treatment are closer to the ages of late 20s and early 30s.

```python
# Scatter plot: Age vs. Treatment
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x="treatment", y="Age", jitter=True, palette="Set2", alpha=0.7)
plt.title("Age vs. Treatment", fontsize=16)
plt.xlabel("Treatment", fontsize=14)
plt.ylabel("Age", fontsize=14)
plt.xticks([0, 1], labels=["Treatment", "No Treatment"])
plt.show()
```

## Scatter Plot:

### Uses a jittered plot to show individual age values for those who received treatment vs. those who didn’t.
### Useful to observe overlapping or concentration of data points.

```python
# Calculate total number of responses
total_responses = len(df)

# Plot the countplot
sns.countplot(data=df, x='treatment')

# Add title and labels
plt.title('Percentage Distribution of Survey Respondents Who Got Treatment')
plt.xlabel('Got Treatment')
plt.ylabel('Percentage')

# Annotate bars with percentages
for bar in plt.gca().patches:
    percentage = f'{(bar.get_height() / total_responses) * 100:.1f}%'  # Calculate percentage
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, percentage,
             ha='center', fontsize=10)

# Show the plot
plt.show()
```
